# Questions
### 1. Which gender per home-ownership group received the most promotions?
### 2. Which gender per wealth group has received the most promotions?
### 3. Which gender per wealth group provides the highest donations?
### 4. Which gender per home-ownership groups provides the highest donations?

```python
dfpv = (donordf.pivot_table(index = 'Gender', columns = 'Homeowner'
                           , values = 'Promotions'
                           , aggfunc = 'mean', fill_value = 0))
dfpv
# Here, we are creating a pivot table featuring gender as the index
# and home-ownershup status as the column headers. The values
# represent the average amound of promotions these indivdiduals
# received. 
```

```python
# Question 1

dplot = dfpv.plot.bar(stacked = True)
labels = ['Female', 'Male']
leglabs = ['No-Home', 'Home']
dplot.legend(leglabs)
dplot.set_ylabel('Average Promotions Sent')
dplot.set_title('Average Promotions Received Based on Gender and Home-Ownership Status')
dplot.set_xticklabels(labels)

print(dplot)

# Assuming that Gender 1 = Male, there are more males that 
# have received promotions than females. Additionally, amongst male 
# recipients, it appears to be an even amount of homeowners and 
# non-homeowners that recieved promotions. . 
```

```python
# Question 2

# Here, we are identifying who recieves the most promotions per
# home-ownership status, gener, and wealth group. 

donordf['Promotions'].groupby([donordf['Homeowner']
                               , donordf['Gender']
                               , donordf['Wealth']]).mean()
# Assuming that Gender 1 = Male, Non-homeowning males, in wealth group
# 6, on average, recieve the most promotions. 
```

```python
# Question 3

# Here, we are identifying the highest donation per gender and
# wealth group. 

donordf['Total_Donation'].groupby([donordf['Gender']
                                  , donordf['Wealth']]).mean()

# Assuming that Gender 1 = Male,on average, males in wealth group 
# 2 give the highest donation.
```

```python
# Question 4

donordf['Total_Donation'].groupby([donordf['Gender']
                                  , donordf['Homeowner']]).mean()

# Assuming that Gender 1 = Male, on average, males that do not own 
# their home give the highest donations. 
```
