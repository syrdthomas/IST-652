## Question 1 & 2

```python
df2pv = (df2.pivot_table(index = 'State', columns = 'Gender'
                           , values = 'Data_Value'
                           , aggfunc = 'sum', fill_value = 0))
# creating pivot table of sum of heart attack counts by gender per state
df2pv
```

```python
df2pv.sort_values(by=('Overall'), ascending=False,inplace=True)
print(df2pv)
# sorting the pivot table from most to least per state and gender
```

```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization

# Let's check the correlation between the variables 
plt.figure(figsize=(20,10)) 
sns.heatmap(df.corr(), annot=True) 
```

```python
# Histogramsof Attributes
# Will not be using this in final report

newdf = df.drop(columns = ['HA_Diag']) 
#dropping heart attack diagnosis from df

fig = plt.figure(figsize=(15, 12))
plt.suptitle('Histograms of Heart Attack Variables', fontsize=20)
for i in range(1, newdf.shape[1] + 1):
    plt.subplot(6, 5, i)
    f = plt.gca()
    f.axes.get_yaxis().set_visible(False)
    f.set_title(newdf.columns.values[i - 1])

    vals = np.size(newdf.iloc[:, i - 1].unique())
    
    plt.hist(newdf.iloc[:, i - 1], bins=vals, color='#3F5D7D')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

## Correlation Matrix

```python
sns.set(style="white")

# Compute the correlation matrix
corr = newdf.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
